```python
import numpy as np
```

## 기본

```python
arr = np.array([1,2,3])
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

## dype 설정

```python
arr = np.array([[1,2,3], [1,2,3]])
print(arr.dtype)
arr2 = np.array([[1,2,3], [2,3,4]], dtype = np.uint8)
print(arr2.dtype)
```

## reshape
* 동일한 size의 다른 shape의 행렬로 바꿔줌
* -1을 인자로 받을 경우 size / 앞에서 나온 shape 값이 들어감

```python
arr = np.array([[1,2,3], [2,4,5]])
print(arr.reshape([3, 2]))
print(arr.reshape(6))
print(arr.reshape([6]))
print(arr.reshape([3, -1]))
```

## random array 생성

```python
arr = np.random.randn(8,8)
print(arr)
print(arr.reshape(-1, 2, 2, 4))
print(arr.reshape(-1))
```

## ravel
* 1차원의 행렬로 쭉 펴주는 함수
* reshape(-1)과 같음

```python
arr = np.random.randn(4,4)
print(arr.ravel())
print(arr.reshape(-1))
arr.ravel() == arr.reshape(-1)
```

## expand_dims
* 값은 유지한 채로 차원만 늘리는 함수

```python
arr = np.random.randn(4,4,4)
print(arr)
print(np.expand_dims(arr, 1))
print(np.expand_dims(arr, -1))
print(np.expand_dims(arr, 0))
```
