```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Basic Attributes

* alpha: 투명도
* kind: 그래프 종류 - 'line', 'bar', 'barh', 'kde', ...
* logy: Y축에 대해 Log scaling
* use_index: 객체의 색인을 눈금 이름으로 사용할지 여부
* rot: 눈금 이름 돌리기 (rotating) 0 ~ 360
* xticks, yticks: x, y축으로 사용할 값
* xlim, ylim: x, y축의 한계값
* grid: 축의 그리드를 표현할지 여부
* subplots: 각 column에 독립된 subplot 그리기
* sharex, sharey: subplots=True이면 같은 x,y축을 공유하고 눈금과 한계를 연결
* figsize: 생성될 그래프의 크기를 tuple로 지정
* title: 그래프의 제목 지정
* legend: subplot의 범례 지정
* sort_columns: column을 알파벳 순서로 그린다


```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

## subplot
* 여러개 그래프 틀 만들기
* 인자는 순서대로 행 수, 열 수, 그래프 번호를 의미함

```python
plt.subplot(3,2,1)
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
```

## Multi Graph 그리기

```python
hist_data = np.random.randn(100).cumsum()
scat_data = np.arange(30)
print(scat_data)
plt.subplot(2, 2, 1)
plt.plot(data)
plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=50) # bins : 막대 하나 당 너비를 결정함. 하나 당 너비 = data range / (bins + 1)
plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.random.rand(30)*30)
plt.show()
```

## 그래프 선 옵션
* 그래프를 그릴 때 표시 되는 색이나 패턴을 바꾸기  
  * 색상: b(파란), g(초록), r(빨강), c(청록), y(노랑), k(검정), w(흰색)
  * 마커: o(원), v(역삼각), ^(삼각), s(네모), +(플러스), .(점)


```python
plt.subplot(2, 2, 1)
plt.plot(data, 'yv')
plt.subplot(2, 2, 2)
plt.plot(data, 'g^')
plt.subplot(2, 2, 3)
plt.plot(data, 'bs')
plt.subplot(2, 2, 4)
plt.plot(data, 'ko')
plt.show()
```
